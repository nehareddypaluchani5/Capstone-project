<h1>Phase 3 : Data Cleaning</h1>

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Country_Region'] = df['Country_Region'].fillna(df['Country_Region'].mode()[0])
df['Year'] = df['Year'].fillna(df['Year'].mean())
df['Genetic_Risk'] = df['Genetic_Risk'].fillna(df['Genetic_Risk'].mean())
df['Air_Pollution'] = df['Air_Pollution'].fillna(df['Air_Pollution'].mean())
df['Alcohol_Use'] = df['Alcohol_Use'].fillna(df['Alcohol_Use'].mean())
df['Smoking'] = df['Smoking'].fillna(df['Smoking'].mean())
df['Obesity_Level'] = df['Obesity_Level'].fillna(df['Obesity_Level'].mean())
df['Treatment_Cost_USD'] = df['Treatment_Cost_USD'].fillna(df['Treatment_Cost_USD'].median())
df['Survival_Years'] = df['Survival_Years'].fillna(df['Survival_Years'].median())
df['Target_Severity_Score'] = df['Target_Severity_Score'].fillna(df['Target_Severity_Score'].median())
df['Cancer_Type'] = df['Cancer_Type'].fillna(df['Cancer_Type'].mode()[0])
df['Cancer_Stage'] = df['Cancer_Stage'].fillna(df['Cancer_Stage'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Age'] = df['Age'].astype(float)
```

```python
df.info()
```

<h1>Phase 5 : Encoding Categorical Variables </h1>

```python
stage_mapping = {
    'STAGE I': 1,
    'STAGE II': 2,
    'STAGE III': 3,
    'STAGE IV': 4
}

df['Cancer_Stage'] = df['Cancer_Stage'].map(stage_mapping)
```

```python
df = pd.get_dummies(df,columns=['Cancer_Type','Gender'], drop_first=True)
```

```python
df.head()
```

```python
df.select_dtypes(include=['int64','float64']).columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Age','Genetic_Risk','Air_Pollution','Alcohol_Use','Smoking','Obesity_Level']] = scaler.fit_transform(df[['Age','Genetic_Risk','Air_Pollution','Alcohol_Use','Smoking','Obesity_Level']])
```

## 📌 Insights & Conclusion

### 🔍 Top 8 Insights from the Data

1. Cancer stage has a major impact on patient survival. Patients in early stages tend to survive longer, while advanced stages show lower survival years.
2. Higher cancer severity is associated with higher treatment costs, indicating that severe cases require more intensive medical care.
3. Genetic risk strongly influences cancer severity, highlighting the importance of family history.
4. Environmental factors such as air pollution contribute to increased cancer severity.
5. Older patients generally show higher severity scores and lower survival years.
6. Lifestyle factors like smoking, alcohol use, and obesity affect cancer severity.
7. Survival years decrease as cancer severity increases.
8. Lung, breast, and colorectal cancers are the most commonly observed cancer types.

### 🏁 Conclusion

This analysis shows that cancer stage is the most critical factor affecting both severity and survival. Genetic, environmental, and lifestyle factors also play significant roles. Early diagnosis and risk management can greatly improve patient outcomes.
